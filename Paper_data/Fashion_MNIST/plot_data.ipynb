```python
import numpy as np
import matplotlib.pyplot as plt

# Load the data from the NumPy file
data = np.load('fashion_data.npy', allow_pickle=True).item()

# Extract the data
train_accuracy = data['train_accuracy']
train_loss = data['train_loss']
test_accuracy = data['test_accuracy']
test_loss = data['test_loss']

# Define the epochs
epochs_train = list(range(len(train_accuracy)))
epochs_test = list(range(len(test_accuracy)))

# Plot training accuracy
plt.figure(figsize=(10, 5))
plt.plot(epochs_train, train_accuracy, label='Training Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy (%)')
plt.title('Training Accuracy over Epochs')
plt.legend()
plt.grid(True)
plt.show()

# Plot training loss
plt.figure(figsize=(10, 5))
plt.plot(epochs_train, train_loss, label='Training Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title('Training Loss over Epochs')
plt.legend()
plt.grid(True)
plt.show()

# Plot test accuracy
plt.figure(figsize=(10, 5))
plt.plot(epochs_test, test_accuracy, label='Test Accuracy')
plt.xlabel('Evaluation Points')
plt.ylabel('Accuracy (%)')
plt.title('Test Accuracy at Evaluation Points')
plt.legend()
plt.grid(True)
plt.show()

# Plot test loss
plt.figure(figsize=(10, 5))
plt.plot(epochs_test, test_loss, label='Test Loss')
plt.xlabel('Evaluation Points')
plt.ylabel('Loss')
plt.title('Test Loss at Evaluation Points')
plt.legend()
plt.grid(True)
plt.show()
```
